```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter
import warnings
warnings.filterwarnings("ignore")
import altair as alt
%matplotlib inline
```

```python
# Import dataset
long_data=pd.read_csv('/content/UserDataNew.csv')
```

```python
long_data.head()
```

```python
# print("***************2022****************")
usage2022=dict()
for i in (long_data.Users.unique()):
    a=long_data[(long_data["Users"]==i)&(long_data["year"]==2022)]
    b=a.Users.sum()
    usage2022.update({i:b})
# print(usage2021)
# print("***************2021****************")
usage2021=dict()
for i in (long_data.Users.unique()):
    a=long_data[(long_data["Users"]==i)&(long_data["year"]==2021)]
    b=a.Users.sum()
    usage2021.update({i:b})
# print(usage2021)
```

```python
table3=pd.DataFrame.from_dict(usage2021,orient="index")
table4=pd.DataFrame.from_dict(usage2022,orient="index")
```

```python
table3.rename(columns={0:'toat usage in 2021'},inplace=True)
table4.rename(columns={0:'toat usage in 2022'},inplace=True)
```

```python
table3.sort_values(by=['toat usage in 2021'],ascending=False,inplace=True)
table4.sort_values(by=['toat usage in 2022'],ascending=False,inplace=True)
```

```python
y2k_pop = long_data.query('year == 2021')
```

```python
alt.Chart(y2k_pop).mark_bar().encode(
    x='Sum of Value',
)
```

```python
alt.Chart(y2k_pop).mark_bar().encode(
    x='Sum of Value',
    y='Users:O',
   
)
```

```python
alt.Chart(y2k_pop).mark_bar().encode(
    x=alt.X('Sum of Value', axis=alt.Axis(title='Sum of Value')),
    y=alt.Y('Users',
        sort=alt.EncodingSortField(field='Sum of Value', order='ascending', op='sum'),
        axis=alt.Axis(title='Users')
    ),
    color=alt.Color('Title:N')
)
```

```python
alt.Chart(y2k_pop).mark_bar().encode(
    x=alt.X('Sum of Value', axis=alt.Axis(title='Sum of Value')),
    y=alt.Y('Users',
        sort=alt.EncodingSortField(field='Sum of Value', order='descending', op='sum'),
        axis=alt.Axis(title='Users')
    ),
    color=alt.Color('Title:N')
)
```
